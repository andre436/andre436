### Transformação de dados

De acordo com o Blog Kondado: "Transformação de dados é o processo de converter a estrutura ou o formato dos seus dados em outro formato ou estrutura em que seja possível analisá-los". Portanto, é necessário que os dados sejam transformados para o formato de número inteiro. Então, os dados da coluna "Região" chamados Norte, Nordeste, Centro-oeste, sudeste e sul serão substituidos respectivamente por 0,1,2,3 e 4; e os dados da coluna "Tem conta" que forem "s" serão mudados para "1" e os "n" para "0" assim como a coluna "Relacionamento".

```python
data['Região'].replace({'Norte':0, 'Nordeste':1, 'Centro-oeste':2, 'Sudeste':3, 'Sul':4}, inplace=True)
data['Tem conta'].replace({'s':1, 'n':0}, inplace=True)
data['Relacionamento'].replace({'s':1, 'n':0}, inplace=True)

data.head()
```

Por fim,a base de dados está pronto para que os algoritmos de machine learning sejam executados. Iniciaremos pelo algoritmo "Árvore de decisão".

## Algoritmo Árvore de decisão.

De acordo com o blog Cronapp: 
                                "A árvore de decisão é um tipo de algoritmo de apoio, cuja dinâmica de funcionamento é baseada                                  em um gráfico ou modelo de decisões alinhado com as suas possíveis consequências. [...] a                                      árvore de decisões é representada por um número mínimo de perguntas. Com base nas respostas,                                    avalia-se a probabilidade de tomar decisões acertadas. [...] O objetivo desse método é abordar                                  a sistemática dos problemas, a fim de que se chegue à conclusão mais lógica possível".
                                
 Então, a seguir aplicaremos o algotitmo e 

```python
y = data.Relacionamento
X = data.drop(columns = ['Relacionamento'])
data_categoricos = pd.get_dummies(data=data, columns=['Tem conta','Relacionamento'])
```

```python
data_categoricos
```

```python
data_categoricos.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_categoricos, y, test_size = 0.30, random_state = 101)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
dtree.fit(X_train, y_train)
```

```python
y_pred_train = dtree.predict(X_train)
y_pred_test = dtree.predict(X_test)
```

### Demonstração da árvore de decisão.

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (16,12))
a = plot_tree(dtree, feature_names = data_categoricos.columns, fontsize = 12, filled = True, 
              class_names = ['Relacionamento não possível', 'relacionamento possível'])
```

### Acurácia

```python
from sklearn.metrics import accuracy_score

print('Acurácia sem dados de treino: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
print('Acurácia sem dados de teste: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```

### Matriz de confusão

```python
from sklearn.metrics import confusion_matrix
mc = confusion_matrix(y_test, y_pred_test)
print('Matriz de confusão\n\n', mc)
```

### Relatório de Classificação

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test))
```

Podemos ver que a classificação feita com a árvore de decisão foi acertiva em 15
